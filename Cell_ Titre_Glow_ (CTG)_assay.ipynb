```python
import pandas as pd
import numpy as np

# Replace 'your_excel_file.xlsx' with the path to your Excel file
excel_file = r"C:\\Users\\User\\Documents\\Python Scripts\\kamini\\varahi_1.0\\kamini_1.xls"

# Load the Excel file into a DataFrame
df = pd.read_excel(excel_file)

# Initialize variables to keep track of the start and end rows
start_row = 3
end_row = start_row + 5  # Gap of 5 rows between datasets

# Initialize a dictionary to store all datasets and their splits
all_datasets = {}

while end_row < len(df):
    # Extract a dataset
    selected_data = df.iloc[start_row:end_row + 1, 3:13]  # Assuming 6x10 matrix

    # Split the dataset into two 3x10 parts
    dataset_3x10_1 = selected_data.iloc[:3, :]
    dataset_3x10_2 = selected_data.iloc[3:, :]

    # Store the original dataset and its splits in the dictionary
    dataset_name = f'Dataset_{len(all_datasets) + 1}'
    all_datasets[dataset_name] = {
        'Original': selected_data,
        'Split_1': dataset_3x10_1,
        'Split_2': dataset_3x10_2
    }

    # Move to the next dataset
    start_row = end_row + 6  # Start of the next dataset
    end_row = start_row + 5  # Gap of 5 rows between datasets
    
    
    
```

```python
# Now you can access all datasets and their splits in different cells using the dictionary 'all_datasets'.
# Example in a new cell:
dataset_name = 'Dataset_1'  # Replace with the specific dataset name you want to access
split_1_data = all_datasets[dataset_name]['Split_1']
split_2_data = all_datasets[dataset_name]['Split_2']


# Example in a new cell:
dataset_name = 'Dataset_3'  # Replace with the specific dataset name you want to access
split_3_data = all_datasets[dataset_name]['Split_1']
split_4_data = all_datasets[dataset_name]['Split_2']

```

```python
# Use split_1_data and split_2_data 

def normalize_split_dataset(split_data):
    # Extract the last column of the split dataset
    last_column = split_data.iloc[:, -1]
    
    # Calculate the mean of the last column
    mean_last_column = last_column.mean()
    
    # Divide each element of the split dataset by the mean
    normalized_data = split_data / mean_last_column
    
    return normalized_data
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
from ipywidgets import interact, IntSlider
from IPython.display import display

# Provided data for q1, q2, q3, and q4
x = list  # Make sure to provide your x values
df1 = q1 
df2 = q2
df3=q3
df4=q4

# Take the logarithm of the x values
log_x = np.log(x)

# Define a nonlinear model function, e.g., a sigmoidal function
def sigmoidal(x, a, b, c, d):
    return a / (1 + np.exp(-b * (x - c))) + d

# Define initial parameter guesses for each dataset
# You can adjust these guesses based on your data and knowledge of the sigmoidal model behavior for each dataset
initial_guesses1 = [1.0, 1.0, 1.0, 1.0]
initial_guesses2 = [1.0, 1.0, 1.0, 1.0]
initial_guesses3 = [1.0, 1.0, 1.0, 1.0]
initial_guesses4 = [1.0, 1.0, 1.0, 1.0]

# Create a custom y-axis formatter to display percentages
percent_formatter = PercentFormatter(xmax=1.0, decimals=1)

# Function to update the plot based on the selected frame
def update_plot(selected_frame):
    fig, ax = plt.subplots(figsize=(7, 6))

    if selected_frame == 1:
        x_data = log_x
        y_data_1 = q1
        y_data_2 = q2
        initial_guesses = initial_guesses1
        title = 'q1 and q2'
    elif selected_frame == 2:
        x_data = log_x
        y_data_1 = q3
        y_data_2 = q4
        initial_guesses = initial_guesses1
        title = 'q3 and q4'
    else:
        x_data = log_x
        y_data_1 = q1
        y_data_2 = q2
        y_data_3 = q3
        y_data_4 = q4
        initial_guesses = initial_guesses1  # Use initial guesses from q1 for all datasets
        title = 'All Datasets'

    x_fit = np.linspace(log_x.min(), log_x.max(), 1000)
    params, _ = curve_fit(sigmoidal, log_x, y_data_1, p0=initial_guesses)
    y_fit_1 = sigmoidal(x_fit, *params)

    params, _ = curve_fit(sigmoidal, log_x, y_data_2, p0=initial_guesses)
    y_fit_2 = sigmoidal(x_fit, *params)

    scatter1 = ax.scatter(x_data, y_data_1, color='red', label='q1')
    scatter2 = ax.scatter(x_data, y_data_2, color='firebrick', label='q2')
    line1 = ax.plot(x_fit, y_fit_1, color='blue', linestyle='--', label='Fitted Curve for q1')
    line2 = ax.plot(x_fit, y_fit_2, color='darkcyan', linestyle='--', label='Fitted Curve for q2')

    if selected_frame == 3:
        params, _ = curve_fit(sigmoidal, log_x, y_data_3, p0=initial_guesses)
        y_fit_3 = sigmoidal(x_fit, *params)

        params, _ = curve_fit(sigmoidal, log_x, y_data_4, p0=initial_guesses)
        y_fit_4 = sigmoidal(x_fit, *params)

        scatter3 = ax.scatter(x_data, y_data_3, color='olive', label='q3')
        scatter4 = ax.scatter(x_data, y_data_4, color='g', label='q4')
        line3 = ax.plot(x_fit, y_fit_3, color='greenyellow', linestyle='--', label='Fitted Curve for q3')
        line4 = ax.plot(x_fit, y_fit_4, color='lime', linestyle='--', label='Fitted Curve for q4')

    ax.set_xlabel('Log(Inhibitor Concentration)')
    ax.set_ylabel('Data (Percentage)')
    ax.yaxis.set_major_formatter(percent_formatter)
    ax.set_title(f'Data and Fitted Curve for {title}')
    ax.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Create a slider widget to select the frame to plot
frame_slider = IntSlider(min=1, max=3, description='Frame:', value=1)
interact(update_plot, selected_frame=frame_slider)
display(frame_slider)
```
